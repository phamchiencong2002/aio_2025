```python
def getting_max_over_kernel(num_list, kernel_size):
    """
    Returns the maximum value in a list over a sliding kernel.
    """
    if kernel_size <= 0:
        raise ValueError("Kernel size must be a pos integer.")
    if kernel_size > len(num_list):
        raise ValueError("Kernel size must be less than or equal to the length of the list.")
    max_values = []
    for i in range(len(num_list) - kernel_size + 1):
        kernel = num_list[i:i+kernel_size]
        max_values.append(max(kernel))
    return max_values
```

```python
def character_counting(s):
    """
    Returns a dictionary with the count of each character in the string.
    """
    char_count = {} # a dictionary to store character counts
    for char in s:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    print(char_count)

character_counting("hello world")
```

```python
def preprocess_text(sentence):
    """
    Preprocesses the input sentence by converting it to lowercase and removing punctuation.
    Parameters:
        sentence (str): Chuỗi câu đầu vào.

    Returns:
        list: Danh sách các từ sau khi xử lý.
    """
    words = []
    sentence = sentence.lower()
    sentence = sentence.strip()
    words = sentence.split()
    return words
print(preprocess_text("I love AI, AI is not eAsY"))
```

```python
def count_word(file_path):
    """
    Counts the frequency of each word in a text file.
    Parameters:
        file_path (str): Đường dẫn đến file
    Returns:
        dict: Từ điển chứa từ và số lần xuất hiện của nó.
    """
    word_count = {}
    with open(file_path, "r") as file:
        for line in file:
            words = preprocess_text(line)
            for word in words:
                if word in word_count:
                    word_count[word] += 1
                else:
                    word_count[word] = 1
    return word_count

file_path = "./P1_data.txt"
word_frequencies = count_word(file_path)
assert word_frequencies.get('who', 0) == 3
print(word_frequencies.get('man', None))
```

```python
def levenshtein_distance(s1, s2):
    """
    Calculates the Levenshtein distance between two strings.
    """
    if len(s1) == 0:
        return len(s2)
    if len(s2) == 0:
        return len(s1)
    if (s1[len(s1) - 1] == s2[len(s2) - 1]):
        return levenshtein_distance(s1[:-1], s2[:-1])
    return 1 + min(levenshtein_distance(s1, s2[:-1]),    # insertion
                    min(levenshtein_distance(s1[:-1], s2),    # Deletion
                    levenshtein_distance(s1[:-1], s2[:-1])  # Substitution
                    ))

levenshtein_distance("anhyeuem", "emyeuanh")
```
