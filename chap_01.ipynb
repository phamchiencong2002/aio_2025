```python
import math
import random
```

```python
def evaluate_f1_components(tp, fp, fn):
    """ tp : true positives >=0, int
        fp : false positives >=0, int
        fn : false negatives >=0, int

        FORMULA:
        precision = tp / (tp + fp)
        recall = tp / (tp + fn)
        f1 = 2 * (precision * recall) / (precision + recall)
    """
    if isinstance(tp, int) and isinstance(fp, int) and isinstance(fn, int):
        if tp < 0 or fp < 0 or fn < 0:
            raise ValueError("tp, fp, and fn must be non-negative integers.")
        precision = tp / (tp +fp) if (tp + fp) > 0 else 0.0
        recall = tp / (tp + fn) if (tp + fn) > 0 else 0.0
        f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0.0
        return precision, recall, f1
    else:
        raise ValueError("tp, fp, and fn must be integers.")

evaluate_f1_components(10, 5, 3)
```

```python
def sigmoid(x):
    """
    Activation function that maps any real value to the (0, 1) interval.
    FORMULA:
    sigmoid(x) = 1 / (1 + exp(-x))
    """
    return 1 / (1 + math.exp(-x))
```

```python
def relu(x):
    """
    Popular activation function in neural networks that outputs the input directly if it is positive; otherwise, it outputs zero.
    """
    return x if (x > 0) else 0
```

```python
def elu(x, alpha=1.0):
    """
    Exponential Linear Unit (ELU) is an activation function that outputs the input directly if it is positive; otherwise, it outputs a scaled exponential function of the input.
    We have to test with different alpha values to get the best results, but 1.0 is a common choice
    """
    return x if (x > 0) else alpha * (math.exp(x) - 1)
```

```python
def is_number(value):
    try:
        float(value)
    except ValueError:
        return False
    return True

def interactive_activation_function():
    x = input("Enter a value: ")
    if is_number(x):
        x = float(x)
    activation = input("Input activation function (sigmoid|relu|elu): ")
    if activation == "sigmoid":
        result = sigmoid(x)
    elif activation == "relu":
        result = relu(x)
    elif activation == "elu":
        alpha_input = input("Enter alpha value for ELU (default is 1.0): ")
        alpha = float(alpha_input) if is_number(alpha_input) else 1.0
        result = elu(x, alpha)
    else:
        print(f"{activation} is invalid activation function. Please choose sigmoid, relu, or elu.")
    return result

interactive_activation_function()
```

```python
def calc_ae(y_target, y_pred):
    """
    Absolute Error (AE) is the absolute difference between the predicted value and the actual target value.
    FORMULA:
    AE = |y_target - y_pred|
    """
    return abs(y_target - y_pred)
```

```python
def calc_se(y_target, y_pred):
    """
    Squared Error (SE) is the square of the difference between the predicted value and the actual target value.
    FORMULA:
    SE = (y_target - y_pred)^2
    """
    return (y_target - y_pred) ** 2
```

```python
def interactive_loss_function():
    num_samples = int(input("Enter the number of samples: "))
    loss_name = input("Input loss function (mae|mse): ")
    final_loss = 0.0
    for i in range(num_samples):
        pred_sample = random.uniform(0, 10)
        target_sample = random.uniform(0, 10)
        if loss_name == 'mae':
            loss = calc_ae(target_sample, pred_sample)
        elif loss_name == 'mse':
            loss = calc_se(target_sample, pred_sample)
        final_loss += loss
        print(f"loss_name : {loss_name}, sample {i}, target: {target_sample:.4f}, prediction: {pred_sample:.4f}, loss: {loss:.4f}")
    final_loss /= num_samples
    if loss_name == 'rmse':
        final_loss = math.sqrt(final_loss)
    print(f"Final {loss_name.upper()}: {final_loss:.4f}")
    return final_loss
interactive_loss_function()
```

```python
def factorial(n):
    """Calculate the factorial of a non-negative integer n.
    FORMULA:
    n! = n * (n-1)! for n > 0
    0! = 1
    """
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers.")
    elif n == 0:
        return 1
    else:
        result = 1
        return result * n * factorial(n - 1)

assert(factorial(5) == 120)
```

```python
def approx_sin(x, n):
    """
    Approximate the sine of x using the Taylor series expansion.
    
    Parametres:
    x (float) : the input angle in radians
    n (int) :  number of terms in the Taylor series expansion
    
    Return:
    float : the approxiamte value of sin(x) using n+1 terms.
    """
    sin_approx = 0.0
    for i in range(n + 1):
        coef = (-1) ** i
        num = x ** (2 * i + 1)
        denom = factorial(2 * i + 1)
        sin_approx += coef * (num / denom)
    return sin_approx

def approx_cos(x,n):
    """
    Approximate the cosine of x using the Taylor series expansion.
    
    Parametres:
    x (float) : the input angle in radians
    n (int) :  number of terms in the Taylor series expansion
    
    Return:
    float : the approxiamte value of cos(x) using n+1 terms.
    """
    cos_approx = 0.0
    for i in range(n + 1):
        coef = (-1) ** i
        num = x ** (2 * i)
        denom = factorial(2 * i)
        cos_approx += coef * (num / denom)
    return cos_approx


def approx_sinh(x, n):
    """
    Approximate the hyperbolic sine of x using the Taylor series expansion.
    
    Parametres:
    x (float) : the input value
    n (int) :  number of terms in the Taylor series expansion
    
    Return:
    float : the approxiamte value of sinh(x) using n+1 terms.
    """
    sinh_approx = 0.0
    for i in range(n + 1):
        coef = 1
        num = x ** (2 * i + 1)
        denom = factorial(2 * i + 1)
        sinh_approx += coef * (num / denom)
    return sinh_approx



def approx_cosh(x, n):
    """
    Approximate the hyperbolic cosine of x using the Taylor series expansion.
    
    Parametres:
    x (float) : the input value
    n (int) :  number of terms in the Taylor series expansion
    
    Return:
    float : the approxiamte value of cosh(x) using n+1 terms.
    """
    cosh_approx = 0.0
    for i in range(n + 1):
        coef = 1
        num = x ** (2 * i)
        denom = factorial(2 * i)
        cosh_approx += coef * (num / denom)
    return cosh_approx

print(approx_sin(x=3.14, n=10))
print(approx_cos(x=3.14, n=10))
print(approx_sinh(x=3.14, n=10))
print(approx_cosh(x=3.14, n=10))
```
